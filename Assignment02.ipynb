```python
import matplotlib.pyplot as plt #importing matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8)) #adjust size of figure
plt.plot(a['Births'], label='Births') #to plot 'Births' and give description to line
plt.plot(a['Deaths'], label='Deaths') #to plot 'Deaths' and give description to line
plt.plot(a['Natural_Increase'], label='Natural_Increase') #to plot 'Natural_Increase' and give description to line
plt.xlabel('Period') #give description to x-axis
plt.ylabel('Data_Count') #give description to y-axis
plt.title('Population Statistics') #give title of the figure
plt.legend() #give description of each line
```
